With a  p-value of 0, the experimental result is within our threshold of significance (alpha = .05). Simulating the null hypothesis using bootstrapping of our sample-signature score data across leprosy-lesion signatures and non-leprosy lesion signatures resulted in results as extreme or more extreme compared to the experimental result 0% of the time simply due to chance. As a result we confidently reject the null hypothesis that there is no difference between leprosy mcacrophage signature scores and non-leprosy macrophage signature scores. After completing 10000 simulations we found that the average difference between leprosy macrophage signature scores and nonleprosy macrophage signature scores to be 0.950838716394661.
