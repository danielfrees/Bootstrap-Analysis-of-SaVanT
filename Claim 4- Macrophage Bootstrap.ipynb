```python
# for DermDB_Erythema nodosum leprosum
nodosum = []
for i in range(len(df.loc['DermDB_Erythema nodosum leprosum'])):
    nodosum.append(float(df.loc['DermDB_Erythema nodosum leprosum'][i]))
    
nodosumScore = sum(nodosum)
print("Total combined Erythema nodosum leprosum score: "+ str(nodosumScore))
print("Average score of Erythema nodosum leprosum score: "+ str(nodosumScore/80)+ '\n')

# for DermDB_Lepromatous leprosy
lepromatous = []
for i in range(len(df.loc['DermDB_Lepromatous leprosy'])):
    lepromatous.append(float(df.loc['DermDB_Lepromatous leprosy'][i]))
    
lepromatousScore = sum(lepromatous)
print("Total combined Lepromatous leprosy score: "+ str(lepromatousScore))
print("Average score Lepromatous leprosy: "+ str(lepromatousScore/80)+ '\n')

# for DermDB_Reversal reaction (leprosy)
reverse = []
for i in range(len(df.loc['DermDB_Reversal reaction (leprosy)'])):
    reverse.append(float(df.loc['DermDB_Reversal reaction (leprosy)'][i]))
    
reversal_rxnScore = sum(reverse)
print("Total combined DermDB_Reversal reaction (leprosy) score: "+ str(reversal_rxnScore))
print("Average score DermDB_Reversal reaction (leprosy): "+ str(reversal_rxnScore/80)+ '\n')
```

```python
# created a new dataframe that excludes the leprosy lesions so that we can compare everything else to it. 
new_df = df.drop(index = ['DermDB_Erythema nodosum leprosum','DermDB_Lepromatous leprosy','DermDB_Reversal reaction (leprosy)'])
```
